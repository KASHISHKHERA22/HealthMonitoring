```python
import pandas as pd
df=pd.read_csv('Training.csv')
```

```python
len(df['prognosis'].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=pd.DataFrame({"Disease":le.fit_transform(df['prognosis'])})
y
newdf=pd.concat([df,y],axis='columns')
newdf

xtrain=newdf.drop(['prognosis','Disease'],axis='columns')
ytrain=newdf['Disease']
y
```

```python
# from sklearn.linear_model import LogisticRegression
# model=LogisticRegression()
# model.fit(xtrain,ytrain)
```

```python
from sklearn.model_selection import train_test_split
Xtrain,xtest,ytrain,ytest= train_test_split(xtrain,ytrain,test_size=0.2,random_state=12)
```

Logisitic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(Xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(ytest, ypred)
print("Accuracy:", accuracy)
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier()
knn_model.fit(Xtrain, ytrain)
```

```python
ypred=knn_model.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(ytest, ypred)
print("Accuracy:", accuracy)
```

SVM

```python
from sklearn.svm import SVC

svm_model = SVC()
svm_model.fit(Xtrain, ytrain)
```
